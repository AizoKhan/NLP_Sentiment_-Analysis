```python
data = pd.read_csv('train_FD002.txt', sep=' ', header=None)
data.head()
```

```python
data = data[[f for f in range(0,26)]]
data.columns = ['ID', 'Cycle', 'OpSet1', 'OpSet2', 'OpSet3', 'SensorMeasure1', 'SensorMeasure2', 'SensorMeasure3', 'SensorMeasure4', 'SensorMeasure5', 'SensorMeasure6', 'SensorMeasure7', 'SensorMeasure8', 'SensorMeasure9', 'SensorMeasure10', 'SensorMeasure11', 'SensorMeasure12', 'SensorMeasure13', 'SensorMeasure14', 'SensorMeasure15', 'SensorMeasure16', 'SensorMeasure17', 'SensorMeasure18', 'SensorMeasure19', 'SensorMeasure20', 'SensorMeasure21']
data
```

```python
data[data['ID'] == 1]
```

Compute outcome variable Remaining Useful Life (RUL)

```python
max_cycles_df = data.groupby(['ID'], sort=False)['Cycle'].max().reset_index().rename(columns={'Cycle':'MaxCycleID'})
max_cycles_df.head()
```

```python
FD001_df = pd.merge(data, max_cycles_df, how='inner', on='ID')
FD001_df['RUL'] = FD001_df['MaxCycleID'] - FD001_df['Cycle']
FD001_df
```

```python
# plot them 
one_engine = []
for i,r in FD001_df.iterrows():
    rul = r['RUL']
    one_engine.append(rul)
    if rul == 0:
        plt.plot(one_engine)
        one_engine = []
        
plt.grid()
```

```python
# As you can see there is no labels in data
plt.scatter(data['ID'],data['Cycle'])
plt.xlabel('Cycles')
plt.ylabel('ID')
plt.show()
```

```python
# KMeans Clustering
data2 = data.loc[:,['Cycle','ID']]
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2)
kmeans.fit(data2)
labels = kmeans.predict(data2)
plt.scatter(data['ID'],data['Cycle'],c = labels)
plt.xlabel('ID')
plt.xlabel('Cycle')
plt.show()
```

```python
# cross tabulation table
df = pd.DataFrame({'labels':labels,"Cycle":data['Cycle']})
ct = pd.crosstab(df['labels'],df['Cycle'])
print(ct)
```
